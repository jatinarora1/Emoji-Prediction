```python
import pandas as pd
train = pd.read_csv("train_emoji.csv",header = None)
test = pd.read_csv("test_emoji.csv",header = None)
```

```python
test.head(5)
```

```python
X_train = train[0]
Y_train = train[1]
X_test = test[0]
Y_test = test[1]
```

```python
for i in range(5):
    print(X_train[i],emoji.emojize(emoji_dictionary[str(Y_train[i])]))
```

```python
f = open("glove.6B.100d.txt",encoding = 'utf-8')
```

```python
Y_test[:5]
```

```python
import numpy as np
embeddings_index= {}
count = 0
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:],dtype = float)
    embeddings_index[word] = coefs
f.close
    
```

```python
emb_dim = embeddings_index["eat"].shape[0]
```

```python
def embeddingOutput(X):
    maxLen = 10
    embedding_out = np.zeros((X.shape[0],maxLen,emb_dim))
    for ix in range(X.shape[0]):
        X[ix] = X[ix].split()
        for ij in range(len(X[ix])):
            try:
                embedding_out[ix][ij] = embeddings_index[X[ix][ij].lower()]
            except:
                embedding_out[ix][ij] = np.zeros((50,))
    return embedding_out
```

```python
embedding_train_matrix = embeddingOutput(X_train)
embedding_test_matrix = embeddingOutput(X_test)
```

```python
from keras.utils import to_categorical
Y_train = to_categorical(Y_train, num_classes=5)
Y_test = to_categorical(Y_test, num_classes=5)
```

```python
print(embedding_train_matrix.shape)
print(embedding_test_matrix.shape)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout, Activation 
```

```python
model = Sequential()
model.add(LSTM(64,input_shape = (10,100)))
model.add(Dropout(0.5))
model.add(Dense(5))
model.add(Activation('softmax'))
model.compile(loss = 'categorical_crossentropy',optimizer = 'adam',metrics = ['acc'])
```

```python
 model.summary()
    
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint
Checkpoint = ModelCheckpoint("best_model.h5", monitor = "val_loss", verbose = 1, save_best_only = True)
Early_Stopping = EarlyStopping(monitor = "val_loss", patience = 10,verbose = 1)
hist = model.fit(embedding_train_matrix,Y_train,epochs = 100, shuffle= True, batch_size=64, validation_split=0.2,callbacks=[Checkpoint,Early_Stopping])
```

```python
model.load_weights("best_model.h5")
```

```python
model.evaluate(embedding_test_matrix,Y_test)
```
